```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread('Smothing/images/haibara.jpg',0)
img= cv2.GaussianBlur(img,(15,15),0)
laplacian = cv2.Laplacian(img,5,cv2.CV_64F)
sobelx = cv2.Sobel(img,cv2.CV_64F,1,0,ksize=9)
sobely = cv2.Sobel(img,cv2.CV_64F,0,1,ksize=9)

imgAdd=np.add(laplacian,img)
imgSub=np.subtract(laplacian,img)
plt.subplot(2,2,1),plt.imshow(img,cmap = 'gray')
plt.title('Original'), plt.xticks([]), plt.yticks([])
plt.subplot(2,2,2),plt.imshow(laplacian,cmap = 'gray')
plt.title('Laplacian'), plt.xticks([]), plt.yticks([])
plt.subplot(2,2,3),plt.imshow(imgAdd,cmap = 'gray')
plt.title('Original-Laplacian'), plt.xticks([]), plt.yticks([])
plt.subplot(2,2,4),plt.imshow(imgSub,cmap = 'gray')
plt.title('Original+Laplacian'), plt.xticks([]), plt.yticks([])


plt.show()

result=np.concatenate((img,laplacian))
result=np.concatenate((result,imgAdd))
result=np.concatenate((result,imgSub))
#cv2.imshow('Flower',result)
#cv2.waitKey(0)
#cv2.destroyAllWindows()
```
