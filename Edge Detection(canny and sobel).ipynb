# Canny Filter

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from scipy import ndimage as ndi
import os
from skimage import io
from skimage import feature


image = io.imread(os.path.join("images", "gun2.tif"),as_grey=True)
img = np.float_(image)
img = ndi.gaussian_filter(img, 4)
img += 0.1*np.random.random(img.shape)


canny1 = feature.canny(img)
canny2 = feature.canny(img, sigma=3)


plt.figure(figsize=(40, 40))

plt.subplot(221),plt.imshow(image,'gray'),plt.title("Original")
plt.xticks([]), plt.yticks([])
plt.subplot(222),plt.imshow(img,'gray'),plt.title("Gaussian Filter")
plt.xticks([]), plt.yticks([])
plt.subplot(223),plt.imshow(canny1,'gray'),plt.title("Canny filter with segma=1")
plt.xticks([]), plt.yticks([])
plt.subplot(224),plt.imshow(canny2,'gray'),plt.title("Canny filter with segma=1")
plt.xticks([]), plt.yticks([])
plt.show()
```

# Laplace Filter and SobelXY

```python
import cv2
from scipy import ndimage
import matplotlib.pyplot as plt
import numpy as np


image = io.imread('Images/gun.tif',0)
img = np.float_(image)
img = ndimage.gaussian_filter(img, 0)

sobelX = cv2.Sobel(img,cv2.CV_64F,1,0,ksize=5)
sobelY = cv2.Sobel(img,cv2.CV_64F,0,1,ksize=5)
laplacian =cv2.Laplacian(img,cv2.CV_64F)



plt.figure(figsize=(30,30))
plt.subplot(221),plt.imshow(image,'gray'),plt.title("Original")
plt.xticks([]), plt.yticks([])
plt.subplot(222),plt.imshow(sobelX,'gray'),plt.title("Sobel X")
plt.xticks([]), plt.yticks([])
plt.subplot(223),plt.imshow(sobelY,'gray'),plt.title("Sobel Y")
plt.xticks([]), plt.yticks([])
plt.subplot(224),plt.imshow(laplacian,'gray'),plt.title("Laplacian")
plt.xticks([]), plt.yticks([])
plt.show()
```
