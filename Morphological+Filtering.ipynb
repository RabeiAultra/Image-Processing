```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
from skimage.morphology import skeletonize, convex_hull_image

image= cv2.imread("images/gun.tif",0)


mask = np.ones((3,3),np.uint8)


erosion = cv2.erode(image,mask)
dilation = cv2.dilate(image,mask)
closing = cv2.morphologyEx(image, cv2.MORPH_CLOSE, mask)
opening = cv2.morphologyEx(image,cv2.MORPH_OPEN, mask)
white_tophat=cv2.morphologyEx(image, cv2.MORPH_TOPHAT, mask)
black_tophat=cv2.morphologyEx(image, cv2.MORPH_BLACKHAT, mask)
skeltonize=skeletonize(image ==0)
convex = convex_hull_image(image ==0)


fig = plt.figure(figsize=(30,30))
filters = [image,erosion,dilation,closing,white_tophat,black_tophat,skeltonize,convex]
for i in range(8):
    plt.subplot(4,3,i+1),plt.imshow(filters[i],cmap = 'gray')
    plt.title("G"), plt.xticks([]), plt.yticks([])
plt.show()
```
