```python
import cv2
import matplotlib.pyplot as plt
from skimage.filters import roberts, sobel


img = cv2.imread('images/mouse.tif',0)

edge_roberts = roberts(img)
edge_sobel = sobel(img)



plt.figure(figsize=(10,10))
plt.subplot(2,2,1),plt.imshow(edge_roberts,'gray'),plt.title('Robert edge detection')

plt.figure(figsize=(10,10))
plt.subplot(2,2,2),plt.imshow(edge_sobel,'gray'),plt.title('Sobel edge detection')


plt.show()
```
